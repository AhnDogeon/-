```python
import pandas as pd
```

```python
mcdonald_data = pd.read_csv('C:/Users/multicampus/yogeuncheo/data_adg/패스트푸드점/맥도날드.txt')
mcdonald_data['dong'] = mcdonald_data.apply(lambda x:x["address"].split()[2], axis=1)
mcdonald_data['gu'] = mcdonald_data.apply(lambda x:x["address"].split()[1], axis=1)
```

```python
mcdonald_data
```

```python
k = 0
for i in mcdonald_data["gu"].value_counts(dropna=False).items():
    print(i)
    k += i[1]
print(k)
```

```python
k = 0
for i in mcdonald_data["dong"].value_counts(dropna=False).items():
    print(i)
    k += i[1]
print(k)
```

```python
korea_data = pd.read_csv('C:/Users/multicampus/Yogeuncheo-PJT/data_gdr/1-2. 서울 법정동 주소목록(최종).txt', sep="\t")
korea_data["goo"] = korea_data.apply(lambda x:x["법정동명"].split()[1] if len(x["법정동명"].split())==3 else "",axis=1)
korea_data["dong"] = korea_data.apply(lambda x:x["법정동명"].split()[2] if len(x["법정동명"].split())==3 else "",axis=1)
korea_data
```

```python
korea_data.rename(columns={"법정동명":"address"}, inplace = True)
```

```python
korea_data
```

```python
new_mc = mcdonald_data.groupby("dong")
new_mc = pd.DataFrame(new_mc["address"].count())
new_mc.columns = ["mac"]
new_korea_data = pd.merge(korea_data,new_mc,on="dong" ,how="left").fillna(0).apply()
```

```python
new_korea_data["mac"] = new_korea_data["mac"].apply(lambda x:int(x))
```
